```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add a bias term to the feature vector
phi = np.c_[X, np.ones_like(X)]

# Solve for coefficients using the normal equation
coefficients = np.linalg.inv(phi.T @ phi) @ phi.T @ y

print("Coefficients:", coefficients)

# Predict the values
yhat = phi @ coefficients

# Plot the data and regression line
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, yhat, label='Regression line', linewidth=2, color='red')
plt.title('Linear Regression using Normal Equation')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Perform linear regression using the least squares method
coefficients, residuals, _, _ = np.linalg.lstsq(np.c_[X, np.ones_like(X)], y, rcond=None)

print("Coefficients:", coefficients)

# Predict the values
yhat = np.c_[X, np.ones_like(X)] @ coefficients

# Plot the data and regression line
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, yhat, label='Regression line', linewidth=2, color='red')
plt.title('Linear Regression using Least Squares')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
